Treating missing values i.e education, cigsPerDay, BPMeds, totChol, BMI, heartrate, glucose

```python
df['education'].fillna(df['education'].mode().iloc[0], inplace=True)
```

```python
df['cigsPerDay'] = df['cigsPerDay'].fillna(df['cigsPerDay'].mean())
```

```python
df['BPMeds'].fillna(df['BPMeds'].mode().iloc[0], inplace=True)
```

```python
df['totChol'] = df['totChol'].fillna(df['totChol'].mean())
```

```python
df['BMI'] = df['BMI'].fillna(df['BMI'].mean())
```

```python
df['heartRate'] = df['heartRate'].fillna(df['heartRate'].mean())
```

```python
df['glucose'] = df['glucose'].fillna(df['glucose'].mean())
```

```python
for col in df_non_binary:
  q1 = df[col].quantile(0.25)
  q3 = df[col].quantile(0.75)
  iqr = q3 - q1
  range_low = q1 - 1.5 * iqr
  range_upper = q3 + 1.5 * iqr
  df = df.loc[(df[col] > range_low) & (df[col] < range_upper)]

df.shape
```

```python
def plot_matrix(data_frame):
    
    columns = df_non_binary.columns
    # Create a figure with a grid of subplots
    fig, axes = plt.subplots(len(columns), len(columns), figsize=(30, 25))
    fig.suptitle('Heartdisease Scatter Plot Matrix', fontsize=20)

    # Iterate over all combinations of columns
    for i, col1 in enumerate(columns):
        for j, col2 in enumerate(columns):
            ax = axes[i, j]

            if i != j:
                # Create scatter plot
                ax.scatter(data_frame[col2], data_frame[col1], alpha=0.5)
                ax.set_xlabel(col2)
                ax.set_ylabel(col1)
            else:
                # Create histogram on the diagonal
                ax.hist(data_frame[col1], bins=30)
                ax.set_xlabel(col1)

            # Remove top and right spines
            ax.spines['top'].set_visible(True)
            ax.spines['right'].set_visible(True)

    plt.tight_layout()
    plt.show()

plot_matrix(df_non_binary)
```

```python
#Split dataset into dependent and independent features
X = df.iloc[:, :-1]
y = df.iloc[:,-1]
```

```python
X.head()
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression()
```

```python
from sklearn.model_selection import GridSearchCV
parameter={'penalty':['l1','l2','elasticnet'],'C':[1,2,3,4,5,6,10,20,23,25,27,30,32,35,38,40,50],'max_iter':[100,200,300]}
```

```python
classifier_regressor = GridSearchCV(classifier, param_grid=parameter, scoring='accuracy',cv = 10)
```

```python
classifier_regressor.fit(X_train, y_train)
```

```python
print(classifier_regressor.best_params_)
```

```python
print(classifier_regressor.best_score_)
```

```python
y_pred = classifier_regressor.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, classification_report
```

```python
score=accuracy_score(y_pred,y_test)
print(score)
```

```python
print(classification_report(y_pred, y_test))
```
